
This visualisation visualizing the relationship between the parental level of education and whether the student has completed a test preparation course using seaborn's countplot. It's a great way to quickly visualize categorical data.

---



```python
# 1 BAR GRAPH
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('clean_data2.csv')

plt.figure(figsize=(10, 6))
sns.countplot(x='parental level of education', hue='test preparation course', data=data, palette='Set3')
plt.title('Parental Education Level and Test Preparation')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Test Preparation', loc='upper right')
plt.show()
```

This is to be plotting a pie chart to visualize the distribution of parental level of education among students. It's a concise way to represent the proportions of different categories within a dataset.

```python
#2 PIE CHART
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('clean_data2.csv')
education_counts = data['parental level of education'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Parental Level of Education')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

This is to be creating a boxplot to visualize the distribution of math, reading, and writing scores among students based on their parental level of education. This visualization will help in understanding the performance of students across different education levels of their parents.

```python
#3 BOX PLOT
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('clean_data2.csv')
parental_order = ["some high school", "high school", "some college", "associate's degree", "bachelor's degree", "master's degree"]
data_melted = pd.melt(data, id_vars=['parental level of education'], value_vars=['math score', 'reading score', 'writing score'],
                      var_name='score_type', value_name='score')
plt.figure(figsize=(12, 8))
sns.boxplot(x='parental level of education', y='score', hue='score_type', data=data_melted, order=parental_order)
plt.title('Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

It is to create a violin plot to visualize the distribution of math scores by gender. Violin plots are great for showing the distribution of numeric data across different categories.

```python
#4 VIOLIN PLOT
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('clean_data2.csv')
plt.figure(figsize=(10, 6))
sns.violinplot(x='gender', y='math score', data=data)
plt.title('Distribution of scores by Gender')
plt.xlabel('Gender')
plt.ylabel('math score')
plt.grid(True)
plt.show()
```

This code generates a line plot to visualize the average math score based on the parental level of education. This visualization can help identify any trends or differences in math performance among students with different parental education levels.

```python
#5 LINE GRAPH
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('clean_data2.csv')
avg_math_scores_by_gender = data.groupby('parental level of education')['math score'].mean()
plt.figure(figsize=(8, 6))
avg_math_scores_by_gender.plot(kind='line', marker='o')
plt.title('Average Math Score by parental level of education')
plt.xlabel('parental level of educationr')
plt.ylabel('Average Math Score')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```
